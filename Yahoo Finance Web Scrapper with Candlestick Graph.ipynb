```python
import pandas as pd
import plotly.graph_objects as pgo
import time
import datetime
```

```python
ticker = input('What stock would you like to see? ')
```

```python
period1 = int(time.mktime(datetime.datetime(2018, 1, 1, 23, 59).timetuple()))
period2 = int(time.mktime(datetime.datetime(2022, 12, 31, 23, 59).timetuple()))
interval = '1d'
```

```python
query_string = f'https://query1.finance.yahoo.com/v7/finance/download/{ticker}?period1={period1}&period2={period2}&interval={interval}&events=history&includeAdjustedClose=true'
```

```python
df = pd.read_csv(query_string)
df.info()
```

```python
fig = pgo.Figure(data = [pgo.Candlestick(x = df['Date'],
                 open = df['Open'],
                 high = df['High'],
                 low = df['Low'],
                 close = df['Close'])])

fig.update_layout(
    width = 900, 
    height = 650,
    title = ticker + ' Stock',
    yaxis_tickprefix = '$', yaxis_tickformat = ',.0f')

fig.update_xaxes(title_text = "Date")
fig.update_yaxes(rangemode="tozero")

fig.show()
```
